# GPFlow Basic Regression
Adapted from GPFlow Tutorials: https://gpflow.readthedocs.io/en/master/notebooks/basics/regression.html \
Using GPFlow to do Gaussian Process Regression on simulated data. \
Firstly, using scipy optimisation (maximising the log marginal likelihood) and then using MCMC with priors on parameters.

```python
import gpflow
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow as tf
import tensorflow_probability as tfp
from tensorflow_probability import distributions as tfd
from gpflow.utilities import print_summary
from gpflow.ci_utils import ci_niter
from gpflow import set_trainable
from gpflow.utilities import print_summary

gpflow.config.set_default_float(np.float64)
gpflow.config.set_default_jitter(1e-4)
gpflow.config.set_default_summary_fmt("notebook")
# convert to float64 for tfp to play nicely with gpflow in 64
f64 = gpflow.utilities.to_default_float

tf.random.set_seed(123)

%matplotlib inline

# The lines below are specific to the notebook format
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 6)
```

Data will be simulated from $f(x)=x sin(x)+\varepsilon$ \
This is a re do of my from scratch gaussian process

```python
sigma_n=0.4

X_train = (np.delete(np.arange(-6,6,0.5), [1,2,3,5,7,8,10,11,12,15])).reshape(-1,1)
epsilon = np.random.normal(loc=0,scale=sigma_n,size=X_train.shape)
Y_train = (X_train*np.sin(X_train)) + epsilon

xs=np.linspace(-10,10,100)

_ = plt.plot(X_train, Y_train, "kx", mew=2)
_ = plt.plot(xs,xs*np.sin(xs) )
```

# Specifying the Kernel

```python
k = gpflow.kernels.RBF()
```

# Create the model

```python
model = gpflow.models.GPR(data=(X_train, Y_train), kernel=k, mean_function=None)
```

```python
print_summary(model)
```

# Optimise Hyperparameters using SciPy 
This maximises the log marginal likelihood to get optimal hyperparameters

```python
optimiser = gpflow.optimizers.Scipy()
opt_logs = optimiser.minimize(model.training_loss, 
                              variables=model.trainable_variables, 
                              options=dict(maxiter=500))
```

```python
print_summary(model)
```

```python
x_s = np.linspace(-10,10,100).reshape(100,1)
mu_s, cov_s = model.predict_f(x_s)

## generate 10 samples from posterior
tf.random.set_seed(1)  # for reproducibility
samples = model.predict_f_samples(x_s, 10)  # shape (10, 100, 1)

## plot
plt.figure(figsize=(12, 6))
plt.plot(X_train, Y_train, "kx", mew=2)
plt.plot(x_s, mu_s, "C0", lw=2)
plt.fill_between(
    x_s[:, 0],
    mu_s[:, 0] - 1.96 * np.sqrt(cov_s[:, 0]),
    mu_s[:, 0] + 1.96 * np.sqrt(cov_s[:, 0]),
    color="C0",
    alpha=0.2,
)

plt.plot(x_s, samples[:, :, 0].numpy().T, "C0", linewidth=0.5)
plt.plot(x_s,x_s*np.sin(x_s),color="orange")
_ = plt.xlim(-10, 10)
```

Here we have plotted the mean of the predicition in dark blue and the confidence interval is shaded. The faint blue lines are samples from our posterior density. The yellow/orange line is the true value of $f(x)$

# MCMC 
GPflow allows you to approximate the posterior over the hyperparemeters after setting a prior for those using Hamiltonian Monte Carlo (HMC)

```python
model.kernel.lengthscales.prior = tfd.Gamma(f64(1.0), f64(1.0))
model.kernel.variance.prior = tfd.Gamma(f64(1.0), f64(1.0))
model.likelihood.variance.prior = tfd.Gamma(f64(1.0), f64(1.0))
```

```python
num_burnin_steps = ci_niter(300)
num_samples = ci_niter(500)

# Note that here we need model.trainable_parameters, not trainable_variables - only parameters can have priors!
hmc_helper = gpflow.optimizers.SamplingHelper(
    model.log_posterior_density, model.trainable_parameters)

hmc = tfp.mcmc.HamiltonianMonteCarlo(
    target_log_prob_fn=hmc_helper.target_log_prob_fn, num_leapfrog_steps=10, step_size=0.01
)
adaptive_hmc = tfp.mcmc.SimpleStepSizeAdaptation(
    hmc, num_adaptation_steps=10, target_accept_prob=f64(0.75), adaptation_rate=0.1
)


@tf.function
def run_chain_fn():
    return tfp.mcmc.sample_chain(
        num_results=num_samples,
        num_burnin_steps=num_burnin_steps,
        current_state=hmc_helper.current_state,
        kernel=adaptive_hmc,
        trace_fn=lambda _, pkr: pkr.inner_results.is_accepted,
    )


samples, traces = run_chain_fn()
parameter_samples = hmc_helper.convert_to_constrained_values(samples)

param_to_name = {param: name for name, param in gpflow.utilities.parameter_dict(model).items()}
```

```python
print_summary(model)
```

```python
def plot_samples(samples, parameters, y_axis_label):
    plt.figure(figsize=(8, 4))
    for val, param in zip(samples, parameters):
        plt.plot(tf.squeeze(val), label=param_to_name[param])
    plt.legend(bbox_to_anchor=(1.0, 1.0))
    plt.xlabel("HMC iteration")
    plt.ylabel(y_axis_label)


plot_samples(samples, model.trainable_parameters, "unconstrained values")
plot_samples(parameter_samples, model.trainable_parameters, "constrained parameter values")
```

```python
def marginal_samples(samples, parameters, y_axis_label):
    fig, axes = plt.subplots(1, len(param_to_name), figsize=(15, 3), constrained_layout=True)
    for ax, val, param in zip(axes, samples, parameters):
        ax.hist(np.stack(val).flatten(), bins=20)
        ax.set_title(param_to_name[param])
    fig.suptitle(y_axis_label)
    plt.show()


marginal_samples(samples, model.trainable_parameters, "unconstrained variable samples")
marginal_samples(parameter_samples, model.trainable_parameters, "constrained parameter samples")
```

```python
plt.figure(figsize=(12, 6))
for i in range(0, num_samples, 20):
    for var, var_samples in zip(hmc_helper.current_state, samples):
        var.assign(var_samples[i])
    f = model.predict_f_samples(x_s,1)
    plt.plot(x_s, f[0, :, :], "C0", lw=2, alpha=0.3)

plt.plot(X_train, Y_train, "kx", mew=2)
_ = plt.xlim(x_s.min(), x_s.max())
_ = plt.ylim(-6, 6)
plt.xlabel("$x$")
plt.ylabel("$f|X,Y$")
plt.title("Posterior GP samples")
plt.show()
```

Comparing this to the same figure in the previous section, you’ll notice that the error bars on the prediction are now substantially larger than before. This is because we are now considering all the predictions that are consistent with the data, not just the “best” prediction. In general, even though it requires much more computation, it is more conservative (and honest) to take all these sources of uncertainty into account.
