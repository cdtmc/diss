```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from numpy.linalg import cholesky, det, lstsq
from scipy.optimize import minimize

sns.set_style(
    style='darkgrid',
    rc={'axes.facecolor':'.9','grid.color':'.8'})

%matplotlib inline
plt.rcParams['figure.figsize'] = [12,6]
plt.rcParams['figure.dpi'] = 100
```

## Kernel

For initial testing only using Squared Exponential Kernel 

$ K_{i,j} = k(x_i,x_j) = \sigma_f^2 \exp\{-\frac{(x_i-x_j)^2}{2 l^2}\} $

This function returns the covariance matrix by computing the square distance matrix associated with the two lists entered.

```python
def kernel(X1, X2, l=0.80, sigma_f=0.80):
    '''
    X1: Array of n points
    X2: Array of m points
    Returns: Covariance matrix (nxm)
    '''
    sqdist = np.sum(X1**2, 1).reshape(-1, 1) + np.sum(X2**2, 1) - 2 * np.dot(X1, X2.T)
    return sigma_f**2 * np.exp(-0.5 / l**2 * sqdist)
```

## Prior
Set our prior mean to be zero and calculate the covariance matrix using $l = 1$ and $\sigma_{f} = 1$.

```python
X = np.arange(-10,10,0.2).reshape(-1,1)

mu = np.zeros(X.shape)
cov = kernel(X,X)

samples = np.random.multivariate_normal(mu.ravel(), cov, 5)

sns.heatmap(cov);
```

## Plotting
Plotting function which plots the mean and 95% confidence interval using the the covariance matrix's diagonal.
Can also plot sample functions and the training data.

```python
def plot_gp(mu,cov,X,X_train=None,Y_train=None,samples=samples):
    X = X.ravel()
    mu = mu.ravel()
    uncertainity = 1.96*np.sqrt(np.diag(cov))
    
    plt.fill_between(X, mu+uncertainity, mu-uncertainity, alpha=0.2, label='95% CI')
    plt.plot(X,mu,label='Mean')
    for i, sample in enumerate(samples):
        plt.plot(X, sample, lw=1, ls='--', label=f'Sample {i+1}')
    if X_train is not None:
        plt.plot(X_train, Y_train,'rx')
    plt.legend()
```

```python
plot_gp(mu,cov,X,samples=samples)
```

## Prediction
The joint distribution of observed data $\boldsymbol{y}$ and predictions $\boldsymbol{f_*}$ is 
$\begin{bmatrix} \boldsymbol{y}\\ \boldsymbol{f_*} \end{bmatrix} \sim \mathcal{N} \left( \boldsymbol{0}, \begin{bmatrix} \boldsymbol{K}_y, \boldsymbol{K}_* \\ \boldsymbol{K}_{*}^T, \boldsymbol{K}_{**} \end{bmatrix}\right). $

The sufficient statistics of the posterior predicitive distribution, $\boldsymbol{\mu_*}$ and $\boldsymbol{\Sigma_*}$, can be computed with:

$\boldsymbol{\mu_*} = \boldsymbol{K}_{*}^T \boldsymbol{K}_y^{-1} \boldsymbol{y},$

$\boldsymbol{\Sigma_*} = \boldsymbol{K}_{**} - \boldsymbol{K}_{*}^T \boldsymbol{K}_y^{-1}\boldsymbol{K}_{*}$

For correctness, the smallest eigenvalue,$e$ of $\boldsymbol{K}_{**}$ should be added to diagonal of $\boldsymbol{K}_{**}$

```python
def posterior_predictive(X_s, X_train, Y_train, l=1.0, sigma_f=1.0, sigma_n=1e-8):
    K = kernel(X_train, X_train, l, sigma_f)+sigma_n**2*np.eye(len(X_train))
    K_s = kernel(X_train,X_s,l,sigma_f)
    K_ss = kernel(X_s,X_s,l,sigma_f)+1e-8*np.eye(len(X_s))
    K_inv= np.linalg.inv(K)
    
    mu_s = K_s.T.dot(K_inv).dot(Y_train)
    cov_s = K_ss - K_s.T.dot(K_inv).dot(K_s)
    
    return mu_s, cov_s
```

# Noise-free Training data
Here we are modelling $y=f(x)$, where $f(x)=x\sin(x)$

```python
# Noise free training data
X_train = (np.delete(np.arange(-6,6,0.5), [1,2,3,5,7,8,10,11,12,15])).reshape(-1,1)
Y_train = X_train*np.sin(X_train)

mu_s, cov_s = posterior_predictive(X,X_train,Y_train)

samples = np.random.multivariate_normal(mu_s.ravel(),cov_s, 5)
plot_gp(mu_s, cov_s, X,X_train=X_train, Y_train=Y_train, samples=samples)
```

Here the covariance matrix, $\boldsymbol{\Sigma_*}$, is shown using colour gradients to show value of the entries. Can see how the noiseless training data changes the diagonal. Can imagine the top left being $x=-10$ on the graph above as move down the diagonal and reach a black spot that is $x=-6$ (first red x) and so on.

```python
sns.heatmap(cov_s,vmin=0);
```

Here is a more illustrative example by only using a few training points and spreading them apart.

```python
X_train = np.concatenate([np.arange(-3,-2,0.5), np.arange(6,7,0.5)], axis=0).reshape(-1,1)
Y_train = X_train*np.sin(X_train)
mu_s, cov_s = posterior_predictive(X, X_train, Y_train)
samples = np.random.multivariate_normal(mu_s.ravel(),cov_s, 5)
plot_gp(mu_s, cov_s, X,X_train=X_train, Y_train=Y_train, samples=samples)
```

```python
sns.heatmap(cov_s,vmin=0.0,vmax=1.0);
```

# Noisy Training

Here we are modelling the problem $y=f(x) + \varepsilon$, where $f(x)=x\sin(x)$ and $\varepsilon \sim \mathcal{N}(0,\sigma_n)$

```python
# Noisy training
sigma_n=0.4

X_train = (np.delete(np.arange(-6,6,0.5), [1,2,3,5,7,8,10,11,12,15])).reshape(-1,1)
epsilon = np.random.normal(loc=0,scale=sigma_n,size=X_train.shape)

Y_train = (X_train*np.sin(X_train)) + epsilon

mu_s, cov_s = posterior_predictive(X, X_train, Y_train, sigma_n=sigma_n)

samples = np.random.multivariate_normal(mu_s.ravel(), cov_s, 5)
plot_gp(mu_s, cov_s, X,X_train=X_train, Y_train=Y_train, samples=samples)
```

Notice the difference between the noise-free covariance heat map and this one. Since introduction of $\varepsilon$ our training points are not exact.

```python
sns.heatmap(cov_s, vmin=0);
```

# Hyperparameters
We have two hyperparameters of the kernel, $l$ and $\sigma_f$.\
$l$ is the localitiy parameter (i.e. distance points influence).\
$\sigma_f$ encodes the amplitude of the fit.\
Shown by the examples below.

```python
hyperparams = [
    (0.2,0.2,0.4),
    (0.2,1.0,0.4),
    (0.2,5.0,0.4),
    (1.0,0.20,0.4),
    (1.0,1.0,0.4),
    (1.0,5.0,0.4),
    (5.0,0.20,0.4),
    (5.0,1.0,0.4),
    (5.0,5.0,0.4),
]

plt.figure(figsize=(12,5))

for i, (l,sigma_f,sigma_n) in enumerate(hyperparams):
    mu_s, cov_s = posterior_predictive(X,X_train,Y_train,l=l,
                                       sigma_f=sigma_f, 
                                       sigma_n=sigma_n)
    plt.subplot(3,3,i+1)
    plt.subplots_adjust(top=2)
    plt.title(f'l={l}, sigma_f = {sigma_f}')
    plot_gp(mu_s, cov_s, X,X_train=X_train, Y_train=Y_train, samples=[])
```

# Hyperparameter tuning
GOAL: Minimise negative log likelihood marginal wrt parameters $l$ and $\sigma_f$ \
Using Algorithm2.1 from Gaussian Processes for Machine Learning

```python
def nll(X_train,Y_train, sigma_n, naive=True):
    Y_train = Y_train.ravel()
    def step_naive(theta):
        K = kernel(X_train, X_train, l=theta[0], sigma_f=theta[1]) + sigma_n**2 * np.eye(len(X_train))
        return 0.5 * np.log(det(K)) + \
               0.5 * Y_train.T.dot(np.linalg.inv(K).dot(Y_train)) + \
               0.5 * len(X_train) * np.log(2*np.pi)
    def step_RW(theta):
        K = kernel(X_train, X_train, l=theta[0], sigma_f=theta[1]) + sigma_n**2 * np.eye(len(X_train))
        L = cholesky(K)
        alpha = lstsq(L.T, lstsq(L, Y_train, rcond=-1)[0],rcond=-1)[0]
        return np.sum(np.log(np.diagonal(L))) + \
               0.5 * Y_train.T.dot(alpha) + \
               0.5 * len(X_train) * np.log(2*np.pi)
    if naive:
        return step_naive
    else:
        return step_RW
    
    
```

```python
res = minimize(nll(X_train, Y_train, sigma_n, naive=False), [1.0, 1.0], bounds=((1e-5, None), (1e-5, None)),
               method='L-BFGS-B')

l_opt, sigma_f_opt = res.x
print(f'optimal l={l_opt}, optimal sigma_f= {sigma_f_opt}')

mu_s, cov_s = posterior_predictive(X,X_train,Y_train,l=l_opt,
                                       sigma_f=sigma_f_opt, 
                                       sigma_n=sigma_n)

plot_gp(mu_s,cov_s,X,X_train=X_train,Y_train=Y_train, samples=[])
```
