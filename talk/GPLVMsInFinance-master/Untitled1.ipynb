```python
import glob
import os
import pandas as pd


# the path to your csv file directory
mycsvdir = 'data'

# get all the csv files in that directory (assuming they have the extension .csv)
csvfiles = glob.glob(os.path.join(mycsvdir, '*.csv'))

# loop through the files and read them in with pandas
dataframes = []  # a list to hold all the individual pandas DataFrames
for csvfile in csvfiles:
    df = pd.read_csv(csvfile,index_col='Date', parse_dates=['Date'])
    coin_str = csvfile[5:]
    coin_str = coin_str[:-4]
    df[coin_str] = df["Close"]
    dataframes.append(df[coin_str])

result = pd.concat(dataframes, axis=1)
```

```python
import numpy as np
portfolio = result.iloc[1230:]
portfolio = portfolio.pct_change()
portfolio = portfolio.iloc[1:]
portfolio.to_csv("crypto_portfolio.csv")
portfolio
```
